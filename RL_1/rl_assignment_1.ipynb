```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def height_of_tank(s0, x, y, z, m, t):
    
    return s0 + ((x - y) / (z - m * m)) * t
```

```python
def voltage_across_cap(V0, C, R, h, t):
    """
        V0: initial voltage
        C: Capacitance
        R: Resistance
        h: timestep width
        t: current time
    """
    r = t / h
    
    val = ((C / h) - (1 / (2 * R))) / ((C / h) + (1 / (2 * R)))
    
    return (val ** r) * V0
```

```python
def queuing_system(lambd, mu, t):
    """
        lambd: parameter for exponential distribution of inter arrival times
        mu: parameters for exponential distribution of serving time
        t: time at which we require the value of number of customers
    """
    
    num_c = 0
    time_curr = 0
    time_arrival = np.random.exponential(lambd)
    time_serve = np.random.exponential(mu)
    
    while True:
        
        if time_curr + min(time_arrival, time_serve) > t:
            break
        
        if time_arrival < time_serve:
            
            if num_c > 0:
                time_serve -= time_arrival
            
            num_c += 1
            time_curr += time_arrival
            time_arrival = np.random.exponential(lambd)
        
        elif time_serve < time_arrival:
            
            time_arrival -= time_serve
            time_curr += time_serve
            
            if num_c > 0:
                # If we have served someone, 
                # then get new serving time
                num_c -= 1
                time_serve = np.random.exponential(mu)
        else:
            
            num_c += 1
            time_curr += time_arrival
            time_arrival = np.random.exponential(lambd)
            
            if num_c > 1:
                
                num_c -= 1
                time_serve = np.random.exponential(mu)
    
    return num_c
```

### Experiment 1

```python
s0 = 5
x = 20
y = 10
z = 1
m = 0
times = np.linspace(0, 10, 21)
heights = height_of_tank(s0, x, y, z, m, times)
plt.plot(times, heights, 'b', label = 'height')
plt.xlabel('time')
plt.ylabel('height')
plt.legend()
plt.show()
```

```python
V0 = 1
C = 1
R = 1
h = 0.0001
times = np.linspace(0, 10, 21)
voltages = voltage_across_cap(V0, C, R, h, times)
plt.plot(times, voltages, 'b', label = 'voltage')
plt.xlabel('time')
plt.ylabel('voltage')
plt.legend()
plt.show()
```

#### μ > λ

```python
lambd = 0.05
mu = 0.07
times = np.linspace(0, 20, 41)
num_customers = []

for t in times:
    num_customers.append(queuing_system(lambd, mu, t))

plt.plot(times, num_customers, 'b', label = 'num_customers')
plt.xlabel('time')
plt.ylabel('number of customers')
plt.legend()
plt.show()    
```
